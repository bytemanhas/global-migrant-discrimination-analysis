```python
import pandas as pd

df = pd.read_csv("south_asian_discrimination_dataset.csv")

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Remove duplicates
df = df.drop_duplicates()

# Standardize column names
df.columns = df.columns.str.lower()

# Convert year to integer
df['year'] = df['year'].astype(int)

df.head()
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Incidents by Country
country_incidents = df.groupby("country")["victims"].sum()

country_incidents.sort_values(ascending=False)
```

```python
country_incidents.plot(kind="bar")
plt.title("Total Incidents by Country")
plt.show()
```

```python
#Incidents Over Time
year_trend = df.groupby("year")["victims"].sum()

year_trend.plot(kind="line", marker="o")
plt.title("Discrimination Incidents Over Time")
plt.show()
```

```python
#Most Common Incident Types
incident_types = df["incident_type"].value_counts()

incident_types.plot(kind="bar")
plt.title("Types of Discrimination Incidents")
plt.show()
```

```python
df.to_csv("cleaned_discrimination_data.csv", index=False)
```
